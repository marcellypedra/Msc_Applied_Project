```python
import pandas as pd
df = pd.read_excel("codebook_codes.xlsx")
df['Code'] = df['Code'].fillna(method='ffill')
grouped_dicts = {}
for code, group in df.groupby('Code'):
    mapping = dict(zip(group['Options'], group['Answer']))
    grouped_dicts[code] = mapping

    for var, mapping in grouped_dicts.items():
        print(f"{var} = {{")
        for key, value in mapping.items():
            print(f"  {key}:\"{value}\",")
        print("}\n")
```
